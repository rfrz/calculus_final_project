```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter
t = 0 # waktu
tmak = 2 * np.pi # waktu akhir
h = 0.00001 # langkah waktunya
N = int((tmak - t) / h) # iterasinya
k = 10 # konstanta pegas
m = 0.1 # massa objek yang terhubung dengan pegas
x = 0.2 # posisi objek
v = 0 # kecepatan
a = -k / m * x  # percepatan

# Arrays for storing results
xp = np.zeros(N + 1)
yp = np.zeros(N + 1)
xp[0] = t
yp[0] = x

# Loop for numerical integration
for i in range(1, N + 1):
    x = x + h * v
    v = v + h * a
    a = -k / m * x
    t = t + h
    xp[i] = t
    yp[i] = x

# Plotting
plt.plot(xp, yp)
plt.grid(True)
plt.title('Gerak pegas')
plt.xlabel('Waktu (t)')
plt.ylabel('Posisi (x)')
plt.show()
```
